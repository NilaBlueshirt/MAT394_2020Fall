## Q3 Function

#### Write a function taking into arbitrary number of name of people (string) and print out a greeting message for each of them

```python
def greeting(*a):
    output = []
    for i in a:
        mesg = "How are you, " + i + "?"
        output.append(mesg)
    return output

#test
greeting('Tom','John','Mary')
```

## Q9 Matplotlib

#### For the random vector `x`, do a bar plot with the index of x as the position, the value of x as the height and width as 0.5

#### Can you mark the largest in the bar plot as red?

```python
import numpy as np
import matplotlib.pyplot as plt 

x=np.random.randn(100)
index=np.arange(len(x),dtype=int)

# This is the index of the largest number
maxind=np.argmax(x)
# plot the maximum in red
for i in index:
    if i != maxind:
        clrs = ['lightblue']
    else:
        clrs = ['red']
        
#clrs = ['lightblue' if (x < np.argmax(x)) else 'red' for i in index ]

# bar plot
plt.bar(index, x, width=0.5, color=clrs)

plt.show()
```

## Q10

#### The following code plots a random walker in a two-dimensional space

#### Can you try a few things below for the code

#### 1. Accelerate the movement of `x`-axis by 10 times

#### 2. Put a statement checking if the absolute value of x is larger than that of y, x does not move and will stay at the previous step

#### 3. Run mutliple times to see if you can get an hourglass shape of trajectory

```python
import numpy as np
import matplotlib.pyplot as plt 

steps=10000
xx=np.zeros(steps)
yy=np.zeros(steps)
for i in range(1,steps):
    xx[i]=xx[i-1]+np.random.randn()*10
    
    yy[i]=yy[i-1]+np.random.randn()
    if np.abs(xx[i])>np.abs(yy[i]):
        xx[i]=xx[i]
    
    
plt.plot(xx,yy,'lightblue')
    
plt.show()
```
