## Q1 

#### Analyze a vector of random numbers with binomial distribution

#### 1. Obtain the mean, standard deviation, n and p of the data assuming you don't know how the samples are generated

#### 2a. draw a normal distribuiton according to the mean and std, with a color of 'r' (this is provided as an example)

#### 2b. draw a binomial distribution according to the n and p, with a color of 'c'

#### 2c. draw a poisson distribution according to the mean, with a color of 'g'


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
from scipy.stats import norm
from scipy.stats import poisson

%matplotlib inline

data=binom.rvs(n=100,p=0.1,size=10000)

plt.hist(data,normed=True,bins=20,color='b',alpha=0.5,label='sample')

# calculate mean, standard deviation, n and p
# binomial distribution
mu=np.mean(data)
sigma=np.std(data)
p=1-sigma**2/mu
n=mu/p

print('mean=', mu, ' sigma=', sigma, ' p=', p, ' n=', n, '\n')

# the variable you can use for all three cases in 2
x=np.arange(30)

# 2a norm
nf=norm(mu,sigma)
plt.plot(x,nf.pdf(x),'r',label='norm')

# 2b binomial
bino=binom(n,p)
plt.plot(x, bino.pmf(x),'c',label='Bionomial pmf')

# 2c poisson 'g'
pois=poisson(mu)
plt.plot(x, pois.pmf(x),'g',label='Poisson pmf')

plt.legend()
plt.show()
```

## Q2

#### Calculate the Pearson, Spearman and Kendall coefficients for the two vectors `x` and `y` below

#### Can you change the last numbers in x and y to make pearsonr and spearmanr differ even more? for example with spearmanr still be 1.0 and pearsonr be close to 0.5?


```python
import numpy as np
import scipy.stats as st

x=[0,1,100]
y=[0,1,1.5]

print(st.pearsonr(x,y))
print(st.spearmanr(x,y))
print(st.kendalltau(x,y))

plt.plot(x,y,'o')
plt.show()
```
