## Q2 Linear equation (1 point)


Solve x, y and z

$ x+ 3y + 5z = 5$

$ 2x + 3y + z = 7$

$ 3x + 4y + z = 10$


1. using system provided function `solve`

2. using inverse of the matrix

3. verify your answers by showing the dot product of `A` and `x` is the same as `b`


```python
# Template
from scipy import linalg
import numpy as np

A = np.array([[1, 3, 5],
              [2, 3, 1],
              [3, 4, 1]])

B = np.array([5, 7, 10])

x = np.linalg.solve(A, B)

print(x)

print(A.dot(x))

print(A.dot(x) == B)
```

# Q3 Least square problem (1 point)

If one additional equation is added, find the least square solution of x, y and z for

$ x+ 3y + 5z = 5$

$ 2x + 3y + z = 7$

$ 3x + 4y + z = 10$

$ 2x + 7y + 3z = 20$

1. using system provided function `lstsq`

2. using pseudo inverse of the matrix


```python
from scipy import linalg
import numpy as np

A = np.array([[1, 3, 5],
              [2, 3, 1],
              [3, 4, 1],
              [2, 7, 3]])

B = np.array([5, 7, 10, 20])

# x is the solution, residuals the sum, rank the matrix rank of input A, and s the singular values of A 
# If B has more than one dimension, lstsq will solve the system corresponding to each column of B
# https://towardsdatascience.com/least-squares-linear-regression-in-python-54b87fc49e77

x, residuals, rank, s = np.linalg.lstsq(A, B, rcond = None)

print(x)
```

## Q4 Eigenvalues (0.5 point)

1. Generate a 5 by 5 random array

2. Obtain the eigenvalues and the eigenvectors of the random array below

```python
import numpy as np
from scipy import linalg

a = np.random.rand(5,5)
vals, vecs = np.linalg.eig(a)

for i, value in enumerate(vals):
    print('Eigenvector:', vecs[:,i], '\n\n', 'Eigenvalue:', value)
```
