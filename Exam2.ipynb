# Q7

#### 1. For the random vectors `x` and `y` below, plot x and y using the `plot` function in `o` notation in `red` without the line

#### 2. Plot a diagonal line with x from `np.min(x)` to `np.max(x)` and y from `np.min(y)` to `np.max(y)` in `k--` notation.



```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline

fig=plt.figure(figsize=[3,3])
x=np.random.randn(100)
y=np.random.randn(100)

plt.plot(x, 'o', color= 'red')
plt.plot(y, 'o', color= 'red')

for x in range(np.mean(x), np.max(x)):
    plt.plot(x, 'k--', color = 'lightblue')
    
for y in range(np.mean(y), np.max(y)):
    plt.plot(y, 'k--', color = 'lightblue')

plt.show()
```

# Extra credit

#### Let's think how we can write a matrix multiplication function with our existing knowledge of Python. 

#### Here is the matrix multiplication algorithm.  Please read the algorithm and finish the code below.



```python
import numpy as np

x=np.random.rand(5,5)
y=np.random.rand(5,3)
print('reference results\n',np.matmul(x,y))

def our_matmul(x,y):
    a = np.zeros((5,3))   
  
    for x0 in x[0, :]:
        for x1 in x[1, :]:
            for x2 in x[2, :]:
                for x3 in x[3, :]:
                    for x4 in x[4, :]:
                        for y0 in y[:, 0]:
                            for y1 in y[:, 1]:
                                for y2 in y[:, 2]:
                                    a[0, 0] = sum(x0 * y0)
                                    a[0, 1] = sum(x0 * y1)
                                    a[0, 2] = sum(x0 * y2)
                                    
                                    a[1, 0] = sum(x1 * y0)
                                    a[1, 1] = sum(x1 * y1)
                                    a[1, 2] = sum(x1 * y2)
                                    
                                    a[2, 0] = sum(x2 * y0)
                                    a[2, 1] = sum(x2 * y1)
                                    a[2, 2] = sum(x2 * y2)
                                    
                                    a[3, 0] = sum(x3 * y0)
                                    a[3, 1] = sum(x3 * y1)
                                    a[3, 2] = sum(x3 * y2)
                                    
                                    a[4, 0] = sum(x4 * y0)
                                    a[4, 1] = sum(x4 * y1)
                                    a[4, 2] = sum(x4 * y2)
    
    
    return a

print('our results\n',our_matmul(x,y))
```
