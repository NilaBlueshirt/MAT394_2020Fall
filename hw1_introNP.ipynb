# Q3

#### The cumulative product of a sequence 1,2,3,4,5 is 1,2,6,24,120.  
#### 1. Can you find a function in NumPy to calculate the cumulative product?
#### 2. And then print out the difference between the cumulative product and cumulative sum?
#### The answer is [0  -1  0  14  105]
#### Note you can learn all the NumPy math function here 
https://docs.scipy.org/doc/numpy-1.13.0/reference/routines.math.html

```python
import numpy as np
a=np.arange(1,6)

# calculate the cumulative product below
# prod_a=?

prod_a = []
prod_a.append(np.cumprod(a))
print(prod_a)


# print cumulative sum below
# sum_a=?

sum_a = []
sum_a.append(np.cumsum(a))
print(sum_a)

# print the difference between product and sum

x = np.array([prod_a, sum_a])
np.diff(x, axis = 0)  # the Difference with axis 0
```

# Q4

#### Can u calculate the value of the function for i upto 100000?
$$ 4\sum_{i=0}^{100000} (-1)^{i} \frac{1}{2i+1}    $$
#### and find the best math constant close to the value

```python
import numpy as np
i=np.arange(100000)

# now calculate the function listed above
# print(np.sum (...))

print(np.sum(4 * ((-1)**i) * (1 / (2*i + 1))))

# print the math constant closest to the value
# print(np.?)

a = np.sum(4 * ((-1)**i) * (1 / (2*i + 1)))
print(np.rint(a))  # rint(x, /[, out, where, casting, order, ]) Round elements of the array to the nearest integer.
```
