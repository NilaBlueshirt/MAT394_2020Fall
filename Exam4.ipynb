### 7.b (0.2 point)

```python
# Q7
from scipy import linalg
import numpy as np

A = np.array([[2, 3, 5],
              [2, 4, 1],
              [3, 3, 1]])

b = np.array([12, 4, 5])

# 7a
x = np.linalg.solve(A, b)
print(x, '\n')

# 7b
x, residuals, rank, s = np.linalg.lstsq(A, b, rcond = None)
print(x)
```

### 8.a (0.8 point)
#### Please obtain the root for the equation below:
$$ x + sin(x) - 3=0 $$

### 8.b (0.2 point)
#### Can you plot $y=x$ and $y=3-sin(x)$ to verify if the crossing point is close to your answer?

```python
# Q8
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# 8.a
# plot the function y = x + sin(x)  3
x = np.linspace(-10, 10, 100)
y = x + np.sin(x) - 3
plt.plot(x, y)
plt.show()

# define the function y = x + sin(x) 5
def func(x):
    return x + np.sin(x) - 3

# solve the function using a good initial guess
res=[]
for x0 in np.arange(-0.5, 0.5, 0.05):
    res.append((x0, fsolve(func, x0 = x0)))
res=np.array(res)

plt.plot(res[:,0], res[:,1], 'b')
plt.show()

# 8.b
# plot the function y = x
x = np.linspace(-10, 10, 100)
y = x 
plt.plot(x, y)

# plot the function y = 3 - sin(x)
x = np.linspace(-10, 10, 100)
y = 3 - np.sin(x)
plt.plot(x, y)

plt.show()
```

## Q9

### 9.a (0.8 point)
#### Try to minimize the equation below:
$$ y= x ^ 2 -5x + 3 $$ 

### 9.b (0.2 point)
#### plot both $ y= x ^ 2 -5x + 3 $  and the minimum point

```python
# Q9
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

x=np.linspace(-1, 10 ,20)

def fun(x):
    return x**2 - 5*x + 3
plt.plot(x, fun(x))

res=minimize(fun, x0=5, method='Nelder-Mead' ) 

x=res.x
print(x)
plt.plot(x, fun(x), 'rx')
plt.show()
```

## Extra credit (1 point)

#### The math 24 game is a card game in which the objective is to find a way to manipulate four intergers so that the end result is 24.

#### The following code below is a simpler version of finding operations between three numbers resulting in 24.
#### Can you change the code so that all the number triples that cannot obtain a result of 24 are printed out?
#### E.g. [1,1,1] does not have any possible operations to obtain 24.

```python
operator = ['+', '-', '*', '/']

goal=24
result = []

for n1 in range(1,10):
    for n2 in range(1,10):
        for n3 in range(1,10):
            # enumerate all possible operators
            active = 1
            for i in operator:
                for j in operator:
                    # enumerate all possible parenthesis
                    res1='(' + str(n1) + i + str(n2) + ')' + j + str(n3)
                    res2= str(n1) + i + '(' + str(n2) + j + str(n3) + ')'
                    try: # deal with invalid operation (divide by zero)
                        if eval(res1)==goal or eval(res2)==goal:
                            active = 0
                    except:
                        break
            if active == 1:
                result.append([n1,n2,n3])
                
result = np.array(result)
goal = result.reshape(len(result),3)
for i in range(goal.shape[0]):
    print(goal[i])
```
